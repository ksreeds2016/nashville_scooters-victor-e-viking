SCOOTERS PER COMPANY BY MONTH - MAY

```python
query3= '''
SELECT companyname, sumdgroup, COUNT(DISTINCT sumdid)
FROM scooters
WHERE sumdgroup ILIKE '%scooter%'
AND pubdatetime BETWEEN '2019-05-01' AND '2019-05-30'
GROUP BY companyname, sumdgroup;
'''

with engine.connect() as connection:    
    scoots_in_May = pd.read_sql(text(query3), con = connection)
```

```python
scoots_in_May.sort_values(by = 'count', ascending = False)
```

```python
scoots_in_May.to_csv('data.csv')
```

AVERAGE DURATION OF SCOOTERS 

```python
query20 = '''
SELECT DISTINCT ROUND(AVG(tripduration), 2) as avg_duration, COUNT(sumdid) as scoots
FROM trips
WHERE startdate BETWEEN '2019-05-01' and '2019-05-31'
AND tripduration >= 1.0 
AND tripduration <= 1440
AND startdate BETWEEN '2019-05-01' and '2019-05-31'
AND tripduration >= 1.0 
AND tripduration <= 1440
'''

with engine.connect() as connection:    
    avg_duration_of_scoots = pd.read_sql(text(query20), con = connection)
    
avg_duration_of_scoots
```

BREAKDOWN OF SCOOTER USAGE BY MORNING, AFTERNOON, EVENING RUSH HOUR, AND NIGHTTIME

Morning - 7-10

```python
query21 = '''
SELECT DISTINCT sumdid, starttime, endtime
FROM trips
WHERE starttime BETWEEN '06:00:00' AND '10:00:00'
AND endtime BETWEEN '06:00:00' AND '10:00:00'
AND startdate BETWEEN '2019-05-01' and '2019-05-31'
AND tripduration >= 1.0 
AND tripduration <= 1440
'''

with engine.connect() as connection:    
    morning_scoots = pd.read_sql(text(query21), con = connection)
    
morning_scoots
```

lunch/afternoonish - 11-2

```python
query22 = '''
SELECT DISTINCT sumdid, starttime, endtime
FROM trips
WHERE starttime BETWEEN '11:00:00' AND '14:00:00'
AND endtime BETWEEN '11:00:00' AND '14:00:00'
'''

with engine.connect() as connection:    
    afternoon_scoots = pd.read_sql(text(query22), con = connection)
    
afternoon_scoots
```

evening/rushhour - 3-6

```python
query23 = '''
SELECT DISTINCT sumdid, starttime, endtime
FROM trips
WHERE starttime BETWEEN '15:00:00' AND '18:00:00'
AND endtime BETWEEN '15:00:00' AND '18:00:00'
'''

with engine.connect() as connection:    
    evening_scoots = pd.read_sql(text(query23), con = connection)
    
evening_scoots
```

nightime/partytime 7-12pm

```python
query24 = '''
SELECT sumdid, starttime, endtime
FROM trips
WHERE starttime BETWEEN '19:00:00' AND '24:00:00'
AND endtime BETWEEN '19:00:00' AND '24:00:00'
AND startdate BETWEEN '2019-05-01' and '2019-05-31'
AND tripduration >= 1.0 
AND tripduration <= 1440
'''

with engine.connect() as connection:    
    nightly_scoots = pd.read_sql(text(query24), con = connection)
    
nightly_scoots
```
